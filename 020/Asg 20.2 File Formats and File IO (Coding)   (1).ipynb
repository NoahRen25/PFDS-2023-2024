<div class="alert alert-block alert-success"><b>Problem 2</b>

Using the text file <mark>studentdata.txt</mark> (shown in Problem 1) write a program that calculates the average grade for each student, and prints out the student’s name along with their average grade.
    
</div>

```python
studentdata = open("C:/Users/nren2_6wfk1d7/Downloads/studentdata.txt", "r")
for i in studentdata:
    value = i.split()
    sum = 0
    j = 1
    while j < len(value):
        sum += int(value[j])
        j +=1
    print(f"{value[0]}'s average grade is: ")
    print(round(sum / (j-1), 2))
```

<div class="alert alert-block alert-success"><b>Problem 3</b>: 

Create a list called <mark>destination</mark> using the data stored in <mark>travel_plans.txt</mark>. Each element of the list should contain a line from the file that lists a country and cities inside that country. Hint: each line that has this information also has a colon <mark>:</mark> in it.<br><p>&nbsp;</p>
    

</div>

```python
travel_plans = open("C:/Users/nren2_6wfk1d7/Downloads/travel_plans.txt", "r")
destination = []
for i in travel_plans:
    value = i.split()
    for j in value:
        if(':' in j.lower()):
            destination.append(value)
print(destination)
```

<div class="alert alert-block alert-success"><b>Problem 4</b>: 

<b>Part One:</b> Write code that stores the contents of the <b>months</b> list in a file named <o>months.txt</i>. Store one month per line.
    
<b>Part Two:</b> Write code that opens the text file that you created in Part One and prints out its contents.
</div>

```python
months = ['January','February','March','April','May','June','July','August','September','October','November','December']
monthsText = open("C:/Users/nren2_6wfk1d7/Downloads/months.txt", 'w')
for i in months:
    monthsText.write(i + ' ')
monthsText.close()
monthsText = open("C:/Users/nren2_6wfk1d7/Downloads/months.txt", 'r')
for i in monthsText:
    value = i.split()
    for j in value:
        print(j)
```

<div class="alert alert-block alert-success"><b>Problem 5:</b> 
    
Create a list called <mark>j_emotions</mark> that contains every word in <mark>emotion_words.txt</mark> that begins with the letter “j”.<br><p>&nbsp;</p>

```python
emotion_words = open("C:/Users/nren2_6wfk1d7/Downloads/emotion_words.txt", "r")
j_emotions = []
for i in emotion_words:
    value = i.split()
    for j in value:
        if('j' in j.lower()):
            j_emotions.append(j)
print(j_emotions)
```
