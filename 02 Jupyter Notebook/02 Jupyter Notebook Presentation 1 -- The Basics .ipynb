#  I.  What is a Jupyter Notebook?

### Answer #1:

● Jupyter Notebook is a web application that allows you to create and share documents that contain live code, equations, visualizations, and explanatory text. It is the application we will be working with in this course.<br>
● Its uses include data cleaning and transformation, numerical simulation, statistical modeling, machine learning, and much more. Jupyter Notebook is similar to a canvas, or an environment, that allows you to use a programming language, in our case Python, to perform computations and to show the results in a very convenient way.<br>
● It is very convenient if you're doing some kind of analytical work, because often you want to include explanatory text, the code that produced the results, and the visualizations, which are shown in Jupyter Notebook. So it is a very convenient way of doing analytical work with any programming language, especially with Python.<br>
● The Jupyter project was born out of the IPython project in 2014. Now, it has evolved to support interactive data science and scientific computing across many other programming languages, so you can use the Jupyter Notebook with many other programming languages (up to almost 20 languages). The name Jupyter comes from [Julia](https://julialang.org/), [Python](https://www.python.org/), and [R](https://www.r-project.org/), which were the three programming languages that were first supported.<br>

### A. Notebook user interfaces


These Jupyter Notebook user interfaces can be very useful as they display important information and notifications while you work on the Notebook.
 
If we go to the **Help** tab and click on the first option, **User Interface Tour**, we can get a quick look at what is involved with these important interfaces of Jupyter Notebook.

<p>&nbsp;</p>

#### 1. Title

This is your filename and this is where you can change the file name of your Notebook.

#### 2. Menu bar

This is similar to any other desktop application; it allows you to do different actions with the Notebook.

#### 3. Toolbar

This is located below the menu bar and contains little icons to perform some commonly used actions ... e.g. saving a file, cutting cells, pasting cells, moving cells, etc.

#### 4. Mode indicator

This is located on the right hand side of the menu bar.  The Notebook has two modes -- Edit mode and Command mode. This will be discussed in greater detail below.

#### 5. Kernal indicator

This display's the status of the system's computation progress.  To interrupt the computations that are in progress, you can use the stop button located on the toolbar.

#### 6. Notification area

This area will display messages, suchas **saving the file**, or **interrupting the kernal**, etc. -- and you will see the actions that you are performing in the notification area.

### C. Keyboard Input Modes 


- A Jupyter Notebook has two modes -- **Edit mode** and **Command mode**.

- Command mode has many shortcut keys that you can use.  In this mode, no icon is displayed in the Mode indicator, and the actions that you can perform in this mode are actions that have to do with the file itself suchas saving the file, copying and pasting cells, etc.

- Edit mode allows you to write code or text in a cell, if you are in a cell.  When you are in Edit mode, you will see a little pencil in the indicator area.

- The border of a selected cell will be GREEN when you are in Edit mode.

- The border of a selected cell will be BLUE when you are in Command mode.

- To go back to Command mode from Edit mode, you can press the Esc key.

- To go back to Edit mode from Command mode, you can press the Enter key.
   
- There are many helpful keyboard shortcuts that you can use and there is a list in the **Help** menu.
