# <span style="color:orange">DATA SCIENCE PROJECT LIFECYCLE STAGES

Embarking on a successful data science project involves a series of well-defined steps, each playing a crucial role in achieving the desired outcome. From defining the problem to evaluating models, these steps provide a comprehensive roadmap for navigating the complex landscape of data-driven problem-solving

- **STEP 1: DEFINE PROBLEM** - This is a crucial step in any data science project as it sets the foundation for the entire project. It is important to define the problem clearly and make sure that it is well-defined and achievable.
- **STEP 2: GATHER DATA** - This step involves gathering the relevant data that will be used for the project. It is important to make sure that the data is of high quality, relevant, and sufficient for the project.
- **STEP 3: PREPARE DATA** - This step involves cleaning, preparing, and performing feature engineering on the data for analysis. Begin by conducting exploratory data analysis (EDA) to gain insights into the data and identify any issues that need to be addressed. Feature engineering, which includes techniques such as standardization, imputation, and encoding, helps create new features or transform existing ones to better represent the problem for predictive models. All the steps involved in data preparation are crucial for a successful analysis.
- **STEP 4: BUILD MODELS** - This step involves selecting and building appropriate models to address the problem defined in Step 1. It is important to choose models that are appropriate for the data and problem, and to experiment with different models to find the best fit.
- **STEP 5: EVALUATE MODELS** - This step involves evaluating the performance of the models built in Step 4. It is important to use appropriate evaluation metrics to determine how well the models are performing and to select the best model for the problem.

# <span style="color:orange">STEP 1: DEFINE PROBLEM

**PROBLEM STATEMENT:**

Predict the price of a car based on various features

**SOLUTIONS OBJECTIVES:**

Using the https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data publicly available dataset from the UCI Machine Learning Repository,
  1. Identify the most important features in predicting car prices
  2. Develop a model with high accuracy in predicting car prices
  
**BRIEF DESCRIPTION OF EACH COLUMN IN DATASET:**

  1.	**symboling:** A measure of the risk associated with insuring a car (categorical: -3, -2, -1, 0, 1, 2, 3).
  2.	**normalized-losses:** Average loss payment per insured vehicle year (continuous).
  3.	**make:** Manufacturer of the car (categorical).
  4.	**fuel-type:** Type of fuel used by the car (categorical: diesel, gas).
  5.	**aspiration:** Type of aspiration system used in the engine (categorical: std, turbo).
  6.	**num-of-doors:** Number of doors on the car (categorical: two, four).
  7.	**body-style:** Body style of the car (categorical: hardtop, wagon, sedan, hatchback, convertible).
  8.	**drive-wheels:** Type of drive wheels the car has (categorical: 4wd, fwd, rwd).
  9.	**engine-location:** Location of the car's engine (categorical: front, rear).
  10.	**wheel-base:** Distance between the center of the front and rear wheels (continuous).
  11.	**length:** Length of the car (continuous).
  12.	**width:** Width of the car (continuous).
  13.	**height:** Height of the car (continuous).
  14.	**curb-weight:** Weight of the car without occupants or baggage (continuous).
  15.	**engine-type:** Type of engine the car has (categorical: dohc, dohcv, l, ohc, ohcf, ohcv, rotor).
  16.	**num-of-cylinders:** Number of cylinders in the car's engine (categorical: eight, five, four, six, three, twelve, two).
  17.	**engine-size:** Size of the car's engine (continuous).
  18.	**fuel-system:** Type of fuel system the car has (categorical: 1bbl, 2bbl, 4bbl, idi, mfi, mpfi, spdi, spfi).
  19.	**bore:** Diameter of the engine cylinders (continuous).
  20.	**stroke:** Distance the piston travels in the cylinder (continuous).
  21.	**compression-ratio:** Ratio of the volume of gas in the cylinder when the piston is at the bottom of its stroke to the             volume of gas when it's at the top of its stroke (continuous).
  22.	**horsepower:** Horsepower of the car's engine (continuous).
  23.	**peak-rpm:** Maximum revolutions per minute of the car's engine (continuous).
  24.	**city-mpg:** Estimated miles per gallon the car gets in city driving (continuous).
  25.	**highway-mpg:** Estimated miles per gallon the car gets on the highway (continuous).
  26.	**price**: Price of the car (continuous).

```python
''' One-hot encode k-1 dummy varibles for categorical variables with k levels (drop_first=True
ensures k-1 dummy variables; this code is a common technique used to preprocess categorical
variables in machine learning models, enabling the use of the categorical variables as inputs to the model) '''

for i in ["make", "fuel-type", "aspiration", "body-style", "drive-wheels", "engine-location", \
    "engine-type", "num-of-cylinders", "fuel-system", "IMP_num-of-doors"] :
    thePrefix = "Z_" + i
    y = pd.get_dummies(df[i], prefix=thePrefix, drop_first=True)   
    df = pd.concat([df, y], axis=1)
df = df.drop(["make", "fuel-type", "aspiration", "body-style", "drive-wheels", "engine-location", \
    "engine-type", "num-of-cylinders", "fuel-system"], axis=1)
```

### MANAGE MISSING NUMERICAL VARIABLES

```python
''' Display median "normalized-losses" by "IMP_num-of-doors" '''

g = df.groupby("IMP_num-of-doors")
i = "normalized-losses"
print(g[i].median())
```

```python
''' Impute median "normalized-losses" by "IMP_num-of-doors" into new field "IMP_normalized-losses"
(the code uses the loc method to assign the median value to the selected rows in the
normalized-losses column). The result is that missing values in the normalized-losses
column where the value in IMP_num-of-doors is two/four will be filled with the median value
of the normalized-losses column for all rows where IMP_num-of-doors is two/four) '''

i = "normalized-losses"
FLAG = "M_" + i
IMP = "IMP_" + i
df[FLAG] = df[i].isna() + 0
df[IMP] = df[i]
df.loc[df[IMP].isna() & df["IMP_num-of-doors"].isin(["four"]), IMP] = \
    df.groupby("IMP_num-of-doors")["normalized-losses"].median()["four"]
df.loc[df[IMP].isna() & df["IMP_num-of-doors"].isin(["two"]), IMP] = \
    df.groupby("IMP_num-of-doors")["normalized-losses"].median()["two"]
df.loc[df[IMP].isna(), IMP] = df[i].median()
numList.remove(i)
```

### DECISION TREE

**A decision tree model is a machine learning method that uses a tree-like structure to make decisions or predictions.** It recursively splits data into subsets based on feature values, creating branches and nodes, with the final nodes (leaves) representing the predicted outcome or class.

**Pros:**

- Easy to understand and visualize
- Handles categorical and numerical data
- Requires minimal data preprocessing

**Cons:**

- Prone to overfitting
- Sensitive to small data changes
- May have suboptimal results compared to other models



[Statquest Video on Decision Trees](https://www.youtube.com/watch?v=_L39rN6gz7Y)


```python
''' Create Decision Tree model '''

dt = DecisionTreeRegressor(max_depth = 8, min_samples_leaf = 5, min_samples_split = 2, random_state = 10)
```

```python
''' Fit Decision Tree model '''

dt.fit(X_train, Y_train)
```

```python
''' Decision Tree train and test R-squared '''

dt_train_r2 = dt.score(X_train, Y_train)
print("Decision Tree Train R-squared:", dt_train_r2)
dt_test_r2 = dt.score(X_test, Y_test)
print("Decision Tree Test R-squared:", dt_test_r2)
```

```python
''' Decision Tree train and test RMSE '''

Y_pred_train = dt.predict(X_train)
Y_pred_train = Y_pred_train.reshape(-1, 1)
Y_train_exp = np.exp(Y_train)
Y_pred_train_exp = np.exp(Y_pred_train)
Y_pred_train_exp = Y_pred_train_exp.reshape(-1, 1)

Y_pred_test = dt.predict(X_test)
Y_pred_test = Y_pred_test.reshape(-1, 1)
Y_test_exp = np.exp(Y_test)
Y_pred_test_exp = np.exp(Y_pred_test)
Y_pred_test_exp = Y_pred_test_exp.reshape(-1, 1)


dt_train_rmse = np.sqrt(mean_squared_error(Y_train_exp, Y_pred_train_exp))
print("Decision Tree Train RMSE:", dt_train_rmse)
dt_test_rmse = np.sqrt(mean_squared_error(Y_test_exp, Y_pred_test_exp))
print("Decision Tree Test RMSE:", dt_test_rmse)
```

```python
''' Print shapes to ensure they are 2D arrays '''''

print(Y_pred_train.shape)
print(Y_train_exp.shape)
print(Y_pred_train_exp.shape)
print(Y_pred_test.shape)
print(Y_test_exp.shape)
print(Y_pred_test_exp.shape)
```

```python
''' Decision Tree train and test MAPE '''

dt_train_mape = np.mean(np.abs((Y_train_exp - Y_pred_train_exp) / Y_train_exp), axis=0) * 100
print("Decision Tree Train MAPE:", dt_train_mape)
dt_test_mape = np.mean(np.abs((Y_test_exp - Y_pred_test_exp) / Y_test_exp), axis=0) * 100
print("Decision Tree Test MAPE:", dt_test_mape)
```

### K-NEAREST NEIGHBORS

**K-nearest neighbors (KNN) is a machine learning algorithm that classifies new data points based on the classification of their K-nearest neighbors in the training set.** It calculates the distance between data points to find the closest K neighbors, and the new data point is classified as the most common class among the K neighbors. KNN is simple, effective, and can be used for classification or regression tasks.

**Pros:**

- Simple and easy to implement
- Works well for small datasets
- Adapts to changes in data

**Cons:**

- Slow for large datasets
- Sensitive to noise and irrelevant features
- Requires feature scaling for distance-based metrics



[Statquest Video on KNN](https://www.youtube.com/watch?v=HVXime0nQeI)

```python
''' Standardize the important features for kNN to:
1. ensure equal contribution of features in distance calculations
2. Prevent features with larger magnitudes from dominating the distance metric
3. Improve the accuracy and performance of the kNN algorithm '''

scaler = StandardScaler()
X_important = X[important_features]
X_train_important, X_test_important = X_important.iloc[train_index], X_important.iloc[test_index]

X_train_important_scaled = scaler.fit_transform(X_train_important)
X_test_important_scaled = scaler.transform(X_test_important)
```

```python
''' Prepare the parameter grid for k (adjust k - below, 31 - as desired)'''

param_grid = {'n_neighbors': np.arange(1, 31)}
```

```python
''' Create a kNN model '''

knn = KNeighborsRegressor()
```

```python
''' Set up the KFold cross-validator (use your preferred number of splits) '''

cv = KFold(n_splits=5, shuffle=True, random_state=42)
```

```python
''' Perform grid search with cross-validation to find the optimal k '''

grid = GridSearchCV(knn, param_grid, cv=cv, scoring='neg_mean_squared_error')
grid.fit(X_train_important_scaled, Y_train)
```

```python
''' Get the optimal k value '''

optimal_k = grid.best_params_['n_neighbors']
print("Optimal k value:", optimal_k)
```

```python
''' Fit the kNN model with the optimal k value '''

knn_optimized = KNeighborsRegressor(n_neighbors=optimal_k)
knn_optimized.fit(X_train_important_scaled, Y_train)
```

```python
''' kNN Train and Test R-squared '''

knn_train_r2 = knn_optimized.score(X_train_important_scaled, Y_train)
print("k-Nearest Neighbors Train R-squared: {:.4f}".format(knn_train_r2))
knn_test_r2 = knn_optimized.score(X_test_important_scaled, Y_test)
print("k-Nearest Neighbors Test R-squared: {:.4f}".format(knn_test_r2))
```

```python
''' kNN Train and Test RMSE '''

Y_pred_train = knn_optimized.predict(X_train_important_scaled)
Y_pred_train = Y_pred_train.reshape(-1, 1)
Y_train_exp = np.exp(Y_train)
Y_pred_train_exp = np.exp(Y_pred_train)
Y_pred_train_exp = Y_pred_train_exp.reshape(-1, 1)

Y_pred_test = knn_optimized.predict(X_test_important_scaled)
Y_pred_test = Y_pred_test.reshape(-1, 1)
Y_test_exp = np.exp(Y_test)
Y_pred_test_exp = np.exp(Y_pred_test)
Y_pred_test_exp = Y_pred_test_exp.reshape(-1, 1)

knn_train_rmse = np.sqrt(mean_squared_error(Y_train_exp, Y_pred_train_exp))
print("kNN Train RMSE: {:.4f}".format(knn_train_rmse))
knn_test_rmse = np.sqrt(mean_squared_error(Y_test_exp, Y_pred_test_exp))
print("kNN Test RMSE: {:.4f}".format(knn_test_rmse))
```

```python
''' kNN Train and Test MAPE '''

knn_train_mape = np.mean(np.abs((Y_train_exp - Y_pred_train_exp) / Y_train_exp), axis=0) * 100
print("k-Nearest Neighbors Train MAPE: {:.2f}%".format(knn_train_mape[0]))
knn_test_mape = np.mean(np.abs((Y_test_exp - Y_pred_test_exp) / Y_test_exp), axis=0) * 100
print("k-Nearest Neighbors Test MAPE: {:.2f}%".format(knn_test_mape[0]))
```
