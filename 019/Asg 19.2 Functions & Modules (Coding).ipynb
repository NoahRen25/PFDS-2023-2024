Write a function <mark>print_triangular_numbers(n)</mark> that prints out the first n triangular numbers. A call to <mark>print_triangular_numbers(5)</mark> would produce the following output:   
    

    
(Hint: use a web search to find out what a triangular number is.)

</div>

```python
def print_triangular_numbers(n):
    sum = 0
    i = 1
    while i in range(n+1):
        print(i, end='     ')
        sum+=i
        print(sum, end='\n')
        i+=1
print_triangular_numbers(5)
```

<div class="alert alert-block alert-success"><b>Problem #4:</b>
Define a function called <mark>common_divisors</mark> that takes two positive integers <mark>m</mark> and <mark>n</mark> as its arguments and returns the list of all common divisors (including 1) of the two integers, unless 1 is the only common divisor. If 1 is the only common divisor, the function should print a statement indicating the two numbers are relatively prime. Otherwise, the function prints the number of common divisors and the list of common divisors.<br>
<p>&nbsp;</p>
Do Parts (a) and (b) below to test your code. 
<p>&nbsp;</p>    
a. Call the function <mark>common_divisors</mark> by passing 5 and 38.
<p>&nbsp;</p> 
b. Call the function <mark>common_divisors</mark> by passing 72 and 48.
<p>&nbsp;</p> 
Your output for  Parts (a) and (b) should look like this:
<p>&nbsp;</p> 
<b>DESIRED OUTPUT</b>



</div>

```python
def common_divisors(m, n):
    bottom = 0
    total = []
    i = 1
    if(m < n):
        bottom = m
    else:
        bottom = n
    while i in range(bottom):
        if m % i == 0 and n % i == 0:
            total.append(i)
        i += 1
    if len(total) == 1:
        return f'{m} and {n} are relatively prime'
    else:
        return total
print(common_divisors(5, 38))
m = 72
n = 48
print(f'{m} and {n} have {len(common_divisors(m, n))}, including 1. \n{common_divisors(m,n)}')
```

```python
def polys(x, a, b, c):
    return a*x**3 + b*x**2 + c*x
a = eval(input('Enter the coefficient of x^3: '))
b = eval(input('Enter the coefficient of x^2: '))
c = eval(input('Enter the coefficient of x: '))
x = 0
min_var = 0
max_var = 0
min = 1000000
max = -1000000
while x < 20:
    x += 0.1
    func = round(polys(x, a, b, c), 2)
    if func < min:
        min = func
        min_var = x
    if func > max:
        max = func
        max_var = x
print(f'The maximum value is {max} when x is {round(max_var, 2)}.')
print(f'The minimum value is {min} when x is {round(min_var, 2)}.')
```

<div class="alert alert-block alert-success"><b>Problem #7:</b> Complete the following steps:<br>
    
One way to calculate how much an investment will be worth is to use the Future Value formula:




(a) Write a function called future_value() which accepts an initial investment I<sub>0</sub> and a number of years n and calculates the future value FV. Include r = 0. 5 as the default yearly rate of return.

(b) Use your future_value() function to calculate the future value of an initial investment of 2000 dollars over 30 years with the default yearly rate of return

(c) Use your future_value() function to calculate the future value of the same initial investment of 2000 dollars over 30 years, but a rate of return of 8% (0.08).

(d) Use your future_value() function to determine when 2000 dollars is invested over 30 years, how much more do you make if the rate of return is 10% (0.10) instead of 5% (0.05).   
<p>&nbsp;</p>
Use formatted print statements to produce meaningful and comprehensive output.

</div>

```python
def future_value(i, n, r=0.05):
    return i*((1 + r)**n)
print(f'With your investment of $2000 over 30 years, with a default 5% rate of return, you will have ${round(future_value(2000, 30), 2)} after 30 years.')
print(f'With your investment of $2000 over 30 years, with a default 8% rate of return, you will have ${round(future_value(2000, 30, 0.08), 2)} after 30 years.')
difference = round(future_value(2000, 30, 0.1), 2) - round(future_value(2000, 30), 2)
print(f'With a investment of $2000 over 30 years, you would get ${round(difference, 2)} more with a 10% rate of return, compared to a 5% rate of return.')
```

Write a function that reverses its string argument.

Test your code on the string 'Python'.

</div>

```python
def reverse(str):
    lis = list(str)
    lists = lis[::-1]
    return ''.join(lists)
print(f'Your reversed string is: {reverse("Python")}')
```

<div class="alert alert-block alert-success"><b>Problem #9:</b> Complete the following steps:<br>

Write a function that will find the average of a variable number of test scores and then print out that average (to 2 decimal places) along with the letter grade that corresponds to that average (see the grade scale below).
    


Test your code using the following test scores:
<p>&nbsp;</p>
scores_term1 = (84, 76, 98, 92, 90)<br>
scores_term2 = (100, 84, 96, 98, 88, 94, 82, 96)

</div>

```python
def avg_score(scores):
    sum = 0
    for i in scores:
        sum += i
    return round(sum / len(scores), 2)
def grade(grade):
    if(grade >= 93):
        return 'A'
    elif(grade >= 85):
        return 'B'
    elif(grade >= 77):
        return 'C'
    elif(grade >= 70):
        return 'D'
    else:
        return 'F'
scores_term1 = (84, 76, 98, 92, 90)
scores_term2 = (100, 84, 96, 98, 88, 94, 82, 96)
grade_term1 = avg_score(scores_term1)
grade_term2 = avg_score(scores_term2)
print(f'In term 1, you averaged a {grade_term1}%, which earned you a {grade(grade_term1)}')
print(f'In term 2, you averaged a {grade_term2}%, which earned you a {grade(grade_term2)}')
```

<div class="alert alert-block alert-success"><b>Problem #10:</b> Complete the following steps:<br>

Write a function that takes a variable number of keyword arguments and then prints out those keyword arguments along with all of the key value pairs using the format key -> value for each key-value pair.
<p>&nbsp;</p>    
Test your code using 3 cases:
<p>&nbsp;</p>
(Teacher1 = 'Mr. Baker', Department = 'Science', Undergrad = 'UW-Madison)<br>
(Teacher2 = 'Ms. Jones', Department = 'Business', Undergrad = 'Marquette', Grad = 'UWM')<br> 
(Teacher3 = 'Mrs. Smith', Department = 'Math', Undergrad = 'MIT', Grad = 'Harvard', PHD = 'Yale', Awards = 'Nobel Peace Prize')

</div>

```python
def prints(**kwargs):
    for key, value in kwargs.items():
        print("{}: {}".format(key, value))
prints(Teacher1 = 'Mr. Baker', Department = 'Science', Undergrad = 'UW-Madison')
print('\n')
prints(Teacher2 = 'Ms. Jones', Department = 'Business', Undergrad = 'Marquette', Grad = 'UWM')
print('\n')
prints(Teacher3 = 'Mrs. Smith', Department = 'Math', Undergrad = 'MIT', Grad = 'Harvard', PHD = 'Yale', Awards = 'Nobel Peace Prize')
```
