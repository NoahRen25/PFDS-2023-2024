<b><font size="4">1. The for Loop</font></b></div>

- A basic building block of all programs is to be able to repeat some code over and over again. In computer science, we refer to this repetitive idea as <b>iteration</b>. In this section, we will explore some mechanisms for basic iteration.

- In Python, the <b>for statement</b> allows us to write programs that implement iteration.

- For example, what if we were curious about the Halloween costumes that our classmates were going to where this year ... we would have several approaches that we could use to ask them

```python
# do the task without using iteration

print('Hey Rohit -- what are you going to be for Halloween this year?')
print('Hey Aniruddh -- what are you going to be for Halloween this year?')
print('Hey Athena -- what are you going to be for Halloween this year?')
print('Hey Cielo -- what are you going to be for Halloween this year?')
print('Hey Adam -- what are you going to be for Halloween this year?')
print('Hey Brandon -- what are you going to be for Halloween this year?')
print('Hey Anubhav -- what are you going to be for Halloween this year?')
```

```python
# do the task using iteration ... use a for loop that iterates over a list

for name in ["Rohit", "Aniruddh", "Athena", "Cielo", "Adam", "Brandon", "Anubhav"]:
    print("Hey", name, "--", "what are you going to be for Halloween this year?")
```

Take a look at the output that will be produced when the code is run. There is one line printed for each friend. Heres how it works:

- <b>name</b> in this for statement is called the loop variable.
- The list of names in the square brackets is called a Python list. 
- Line 2 is the loop body. The loop body is always indented. The indentation determines exactly what statements are in the loop. The loop body is performed one time for each name in the list.
- On each iteration or pass of the loop, a check is done to see if there are still more items to be processed. If there are none left (this is called the terminating condition of the loop), the loop has finished. Program execution continues at the next statement after the loop body.
- If there are items still to be processed, the loop variable is updated to refer to the next item in the list. This means, in this case, that the loop body is executed here 7 times, and each time name will refer to a different friend.
- At the end of each execution of the body of the loop, Python returns to the for statement, to see if there are more items to be handled.

<b><font size="4">3. The Range() Function</font></b></div>

- It turns out that generating lists with a specific number of integers is a very common thing to do, especially when you want to write simple <b>for loop</b> controlled iteration. 

- Even though you can use any four items, or any four integers for that matter, the conventional thing to do is to use a list of integers starting with 0. 

- In fact, these lists are so popular that Python gives us special built-in range objects that can deliver a sequence of values to the for loop. 

- When called with one parameter, the sequence provided by range always starts with 0. If you ask for range(4), then you will get 4 values starting with 0. In other words, 0, 1, 2, and finally 3. Notice that 4 is not included since we started with 0. Likewise, range(10) provides 10 values, 0 through 9.



<b>Note</b><br>

Computer scientists like to count from 0.

```python
# So to repeat something five times, a good Python programmer would do this ...

for i in range(5):
    print(i ** 3)
```

- The range function is actually a very powerful function when it comes to creating sequences of integers. 
- It can take one, two, or three parameters. 
- We have seen the simplest case of one parameter such as range(5) which creates [0, 1, 2, 3, 4]. But what if we really want to have the sequence [1, 2, 3, 4, 5]? 
- We can do this by using a two parameter version of range where the first parameter is the starting point and the second parameter is the ending point. The evaluation of range(1,6) produces the desired sequence. 
- What happened to the 6? In this case we interpret the parameters of the range function to mean range(start,beyondLast), where beyondLast means an index past the last index we want. In the 2-parameter version of range, that is the last index desired + 1.

```python
for i in range(1,6):
    print(i)
```

```python
# the range() function is lazy 

range(5)
print(range(5))
list(range(5))
```

- Finally, suppose we want to have a sequence of odd numbers. How would we do that? 
- Easy, we add another parameter, a step, that tells range what to count by. 
- For odd numbers we want to start at 1 and count by 2s. So if we wanted the first 10 odd numbers we would use range(1,20,2). - The most general form of the range is <b>range(start, beyondLast, step)</b>. 
- You can also create a sequence of numbers that starts big and gets smaller by using a negative value for the step parameter.

```python
# range() function examples

print(list(range(7)))
print()

print(list(range(3, 11)))
print()

print(list(range(1, 20, 2)))
print()
      
print(list(range(10, 0, -1)))
print()

print(list(range(25, 0, -3)))
print()

print(list(range(2, 21, 4)))
print()

print(list(range(2, 23, 4)))
print()
```

<b><font size="4">6. The while Statement</font></b></div>

- There is another Python statement that can also be used to build an iteration. 
- It is called the <b>while statement</b>. 
- The while statement provides a much more general mechanism for iterating. 
- Similar to the if statement, it uses a boolean expression to control the flow of execution. 
- The body of <b>while</b> will be repeated as long as the controlling boolean expression evaluates to True.

The following figure shows the flow of control for a while statement.



- We can use the while loop to create any type of iteration we wish, including anything that we have previously done with a for loop.
- See the examples below, which show how to use both a for loop and a while statement to peform the same iteration.
- Instead of relying on the range function to produce the numbers for our summation, we will need to produce them ourselves. 
- To to this, we will create a variable called number and <b>initialize</b> it to 1, the first number in the summation. 
- Every iteration will add number to the running total sum until all the values have been used. 
- In order to control the iteration, we must create a boolean expression that evaluates to True as long as we want to keep adding values to our running total. 
- In this case, as long as number is less than or equal to the bound, we should keep going.

```python
# Add the first ten natural numbers ... use a for loop

bound = 10
sum = 0
for number in range(1, bound + 1):
    sum = sum + number           
    
print(sum)
```

```python
# Add the first ten natural numbers ... use a while statement

bound = 10
sum  = 0
number = 1
while number <= bound:
    sum = sum + number
    number = number + 1

print(sum)
```

- You can almost read the while statement as if it were in natural language. 
- It means, while <b>number</b> is less than or equal to <b>bound</b>, continue executing the body of the loop. 
- Within the body, each time, update <b>sum</b> using an accumulator pattern and increment <b>number</b>. 
- After the body of the loop, we go back up to the condition of the while and reevaluate it. 
- When <b>number</b> becomes greater than <b>bound</b>, the condition fails and flow of control continues to the next statement after the body of the loop.

Use the [Python Tutor -- Visualize Code Execution](https://pythontutor.com/visualize.html#mode=edit) tool to visualize the while statement for the code below.

More formally, here is the flow of execution for a while statement:

Evaluate the condition, yielding False or True.

1. If the condition is False, exit the while statement and continue execution at the next statement.

2. If the condition is True, execute each of the statements in the body and then go back to step 1.

3. The body consists of all of the statements below the header with the same indentation.

- This type of flow is called a loop because the third step loops back around to the top. 
- Notice that if the condition is False the first time through the loop, the statements inside the loop are never executed.

- The body of the loop should change the value of one or more variables so that eventually the condition becomes False and the loop terminates. 
- Otherwise the loop will repeat forever. This is called an <b>infinite loop</b>. An endless source of amusement for computer scientists is the observation that the directions written on the back of the shampoo bottle (lather, rinse, repeat) create an infinite loop.

In the case shown above, we can prove that the loop terminates because we know that the value of <b>bound</b> is finite, and we can see that the value of <b>number</b> increments each time through the loop, so eventually it will have to exceed <b>bound</b>. In other cases, it is not so easy to tell.

<b>Note</b>

Introduction of the while statement causes us to think about the types of iteration we have seen. The for statement will always iterate through a sequence of values like the list of names for the party or the list of numbers created by range. Since we know that it will iterate once for each value in the collection, it is often said that a for loop creates a <b>definite iteration</b> because we definitely know how many times we are going to iterate. On the other hand, the while statement is dependent on a condition that needs to evaluate to False in order for the loop to terminate. Since we do not necessarily know when this will happen, it creates what we call <b>indefinite iteration</b>. Indefinite iteration simply means that we dont know how many times we will repeat but eventually the condition controlling the iteration will fail and the iteration will stop. (Unless we have an infinite loop which is of course a problem.)

- What you will notice here is that the while loop is more work for you  the programmer  than the equivalent for loop. 
- When using a while loop you have to control the loop variable yourself. You give it an initial value, test for completion, and then make sure you change something in the body so that the loop terminates.<br>
<p>&nbsp;</p>
- So why have two kinds of loop if for looks easier? 
- Answer: Because in situations that involve indefinite iteration we need the extra power that we get from the while loop.

<b><font size="4">7. The 3n + 1 Sequence</font></b></div>

- As another example of indefinite iteration, lets look at a sequence that has fascinated mathematicians for many years. 
- The rule for creating the sequence is to start from some positive integer, call it n, and to generate the next term of the sequence from n, either by halving n, whenever n is even, or else by multiplying it by three and adding 1 when it is odd. 
- The sequence terminates when n reaches 1.

```python
n = int(input("Choose a positive integer:  "))

print(n)
counter = 0

while n != 1: 
    if n % 2 == 0:            # n is even
        n = n // 2
    else:                     # n is odd
         n = n * 3 + 1
    counter = counter + 1
    print(n)                  # the last print is 1

print()
print("The number of iterations needed to reach termination was", counter,".")
```

- The condition for this loop is n != 1. The loop will continue running until n == 1 (which will make the condition false).

- Each time through the loop, the program prints the value of n and then checks whether it is even or odd using the remainder operator. If it is even, the value of n is divided by 2 using integer division. If it is odd, the value is replaced by n * 3 + 1. 

- Since n sometimes increases and sometimes decreases, there is no obvious proof that n will ever reach 1, or that the program terminates. For some particular values of n, we can prove termination. For example, if the starting value is a power of two, then the value of n will be even each time through the loop until it reaches 1.

- You might like to have some fun and see if you can find a small starting number that needs more than a hundred steps before it terminates.

***

- Particular values aside, the interesting question is whether we can prove that this sequence terminates for all positive values of n. So far, no one has been able to prove it or disprove it!

- Think carefully about what would be needed for a proof or disproof of the hypothesis All positive integers will eventually converge to 1. With fast computers we have been able to test every integer up to very large values, and so far, they all eventually end up at 1. But this doesnt mean that there might not be some as-yet untested number which does not reduce to 1.

- Youll notice that if you dont stop when you reach one, the sequence gets into its own loop: 1, 4, 2, 1, 4, 2, 1, 4, and so on. One possibility is that there might be other cycles that we just havent found.

<b><font size="4">8. Traversal and the for Loop: By Item</font></b></div>

- A lot of computations involve processing a collection one item at a time. 
- For strings this means that we would like to process one character at a time. 
- Often we start at the beginning, select each character in turn, do something to it, and continue until the end. 
- This pattern of processing is called a <b>traversal</b>
- Since a string is simply a sequence of characters, the for loop iterates over each character automatically.

```python
# Using a for loop to iterate over each character

for char in "Engage, Ask, Listen, Connect":
    print(char)
```

- The loop variable <b>char</b> is automatically reassigned each character in the string Engage, Ask, Listen, Connect. 
- We will refer to this type of sequence iteration as <b>iteration by item</b>
- Note that it is only possible to process the characters one at a time from left to right.

<b><font size="4">9. Traversal and the for Loop: By Index</font></b></div>

- It is also possible to use the range function to systematically generate the indices of the characters. 
- The for loop can then be used to iterate over these positions. 
- These positions can be used together with the indexing operator to access the individual characters in the string.

```python
# Using the range() function to generates indices for the characters in a string

sport = "football"
for index in range(8):
    current_char = sport[index]
    print(current_char)
```

- The index positions in football are 0,1,2,3, 4, 5, 6, and 7. 
- This is exactly the same sequence of integers returned by range(8). 
- The first time through the for loop, <b>index</b> will be 0 and the f will be printed. Then, <b>index</b> will be reassigned to 1 and o will be displayed. This will repeat for all the range values up to but not including 8. Since l has index 7, this will be exactly right to show all of the characters.

***

- In order to make the iteration more general, the <b>len function</b> can be used to provide the bound for range. 
- This is a very common pattern for traversing any sequence by position. 
- Make sure you understand why the range() function behaves correctly when using len of the string as its parameter value.

```python
# Using len of the string as the range() function parameter value.

sport = "basketball"
for index in range(len(sport)):
    print(sport[index])
```

- You may also note that iteration by position / index allows the programmer to control the direction of the traversal by changing the sequence of index values. 
- Recall that we can create ranges that count down as well as up so the following code will print the characters from right to left.

```python
# Iteration by position / index ... 

sport = "football"
for index in range(len(sport)-1, -1, -1):
    print(sport[index])
```

<b><font size="4">10. Traversal and the while Loop</font></b></div>

- The while loop can also control the generation of the index values. 
- Remember that the programmer is responsible for setting up the initial condition, making sure that the condition is correct, and making sure that something changes inside the body to guarantee that the condition will eventually fail.

```python
# Using a while loop to generate index values

sport = "football"

index = 0
while index < len(sport):
    print(sport[index])
    index = index + 1
```

- The loop condition is index < len(sport), so when index is equal to the length of the string, the condition is false, and the body of the loop is not executed. 
- The last character accessed is the one with the index len(sport)-1, which is the last character in the string.

<b><font size="4">11. Lists and for Loop</font></b></div>

- It is also possible to perform <b>list traversal</b> using iteration by item as well as iteration by index.

```python
# list traversal ... iteration by item

directions = ["North", "South", "East", "West"]

for adirection in directions:     # by item
    print(adirection)
```

- It almost reads like natural language: For (every) direction in (the list of) directions, print (the name of the) direction.
<p>&nbsp;</p>
- We can also use the indices to access the items in an iterative fashion.

```python
# list traversal ... iteration by index

directions = ["North", "South", "East", "West"]

for index in range(len(directions)):     # by index
    print(directions[index])
```

- In this example, each time through the loop, the variable <b>index</b> is used as an index into the list, printing the corresponding element. 
- Note that we used len as the upper bound on the range so that we can iterate correctly no matter how many items are in the list.

### Any sequence expression can be used in a for loop. See the examples below:

```python
# Using the range() funvtion to create a sequence of integers (to print out multiples of 5 between 0 and 40)

for number in range(41):
    if number % 5 == 0:
        print(number)
```

- Since lists are mutable, it is often desirable to traverse a list, modifying each of its elements as you go. 
- The following code cubes all the numbers from 1 to 7 using iteration by position.

```python
# Using the range() and len() function to iterate by position

numbers = [1, 2, 3, 4, 5, 6, 7]

for i in range(len(numbers)):
    numbers[i] = numbers[i] ** 3

print(numbers)
```

- In Python, an easy way to create a list from a sequence of values based on some selection criteriai is to use a <b>list comprehension</b>. 
- List comprehensions are concise ways to create lists. The general syntax is: 

<br> ... where the if clause is optional.

```python
# list comprehension

startlist = [1,2, 3, 4, 5, 6, 7]

endlist = [item ** 3 for item in startlist]

print(endlist)
```

- The <b>expression</b> describes each element of the list that is being built. 
- The <b>for clause</b> iterates through each item in a sequence. 
- The items are filtered by the <b>if clause</b> if there is one. 
- In the example above, the for statement lets item take on all the values in the list startlist. Each item is then cubed before it is added to the list that is being built. The result is a list of cubes of the values in endlist.
