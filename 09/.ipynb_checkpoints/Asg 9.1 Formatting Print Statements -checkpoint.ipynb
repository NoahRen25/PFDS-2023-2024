<div class="alert alert-block alert-success"><b>Question #8:</b> 

<b>Step #1:</b> Write code that asks the user for their dog's name and then asks the user for their dog's age in years.
<p>&nbsp;</p>    
<b>Step #2:</b> Write a formatted print statement to print the dog's name and age in dog years (to the nearest tenth).
<p>&nbsp;</p>    
You will need to use the formula below to calculate the age of the dog in dog years.
<p>&nbsp;</p   
<center>dog_years = (10.5 * 2) + (4 * (age - 2))</center>
<p>&nbsp;</p>    
Run your code for a dog named Charlie that is 10.5 years old.  If you have written the proper code, you should get the following output:
<p>&nbsp;</p>



</div>

```python
name = input("What is your dog's name? ")
age = float(input("How old (in years) is your dog? "))
dog_years = (10.5 * 2) + (4 * (age - 2))
print(f'{name} is {dog_years} years old in dog years.')
```
