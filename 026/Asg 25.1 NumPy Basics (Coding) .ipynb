<div class="alert alert-block alert-success"><b>Problem #1:</b>

Create and print out a 3 x 4 integer array and use formatted print statements to print out the following attributes for the array that you created.

- size
- shape
- number of dimensions
- data type
- the length of each element of array in bytes

<p>&nbsp;</p>
    <center><u>Sample Output</u></center><br>



</div>

```python
np_2d = np.array([[5, 3, 7, 8],[2, 9, 1, 1], [4, 8, 9, 2]])
print(np_2d)
print()
print('NumPy Array Attributes: ')
print(f'Array size: {np_2d.size}')
print(f'Array shape: {np_2d.shape}')
print(f'Number of dimensions in array: {np_2d.ndim}')
print(f'Array data type: {np_2d.dtype}')
print(f'The byte length of each element: {np_2d.nbytes / np_2d.size}')
```

<div class="alert alert-block alert-success"><b>Problem #4:</b>

In the following code cell you will be provided with a NumPy array. Use that array to complete the following tasks:

a. Return a array that contains the second column.<br>
b. Return an array that contains the second through fourth row (including the fourth row).<br>
c. Return an array that contains the elements found in the third through fifth row (including the fifth row) and the second through fifth column (including the fifth column)

<p>&nbsp;</p>
<u>Desired Output</u><br>



</div>

```python
array = np.array([[1, 5, 7, 9, 0, 4], [7, 7, 7, 2, 0, 8], [4, 2, 3, 9, 1, 1], [2, 4, 3, 5, 5, 5], [0, 6, 6, 2, 1, 7]])
print(array)
print('Printing array of items in the second column from all rows')
print(array[:, 1])
print('Printing array of items from the second through fourth row')
print(array[1:4, :])
print('Printing array of items from the third through fifth row and the second through fifth column')
print(array[2:5, 1:5])
```
