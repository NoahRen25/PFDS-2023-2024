<div class="alert alert-block alert-success"><b>Problem #2:</b>
    
Sort the given NumPy array in the following ways:<br>
<p>&nbsp;</p>  
<b>initial_array = numpy.array([[41,26,31,57],[73,65,19, 89],[61,12,9, 29], [73, 71, 98, 11]])</b><br>
<p>&nbsp;</p> 
a. Sort the flattened array<br>    
b. Sort in ascending order and return a 4 x 4 array<br>
c. Sort the columns<br>
d. Sort the rows<br>
e. Sort by setting the axis = -1<br>
f. Sort array by the third row<br>
g. Sort the array by the fourth column<br>
h. Sort the columns in descending order<br>
i. Sort the rows in descending order<br>
j. Sort in reverse/descending order and return a 4 x 4 array<br>
<p>&nbsp;</p>
<center><u>Desired Output</center></u><br


</div>
   

For Help:<br>
 
For questions f and g ... the links below may help

[Sort NumPy Arrays by Columns or Rows](https://opensourceoptions.com/blog/sort-numpy-arrays-by-columns-or-rows/)

[NumPy Sort Array by Column With the numpy.argsort](https://www.delftstack.com/howto/numpy/python-numpy-sort-array-by-column/)

```python
print('Printing the initial array: ')
initial_array = np.array([[41,26,31,57],[73,65,19, 89],[61,12,9, 29], [73, 71, 98, 11]])
print('a. Sorting the flattened array: ')
flat = initial_array.flatten()
sort = np.sort(flat)
print(sort)
print('b. Sorting original array in ascending order: ')
reshape = sort.reshape(4, 4)
print(reshape)
print('c. Sorting original array by columns: ')
column = np.sort(initial_array, 0)
print(column)
print('d. Sorting original array by rows: ')
row = np.sort(initial_array, 1)
print(row)
print('e. Sorting original array setting axis = -1: ')
sorte = np.sort(initial_array, -1)
print(sorte)
print('f. Sorting original array by 3rd row: ')
f = initial_array
f = f[:, f[2, :].argsort()]
print(f)
print('g. Sorting original array by 4th column: ')
g = initial_array
g = g[g[:, 3].argsort()]
print(g)
print('h. Sort the columns in descending order')
h = -np.sort(-initial_array, 0)
print(h)
print('i. Sort the rows in descending order')
i = -np.sort(-initial_array, 1)
print(i)
print('j. Sort in reverse/descending order and return a 4 x 4 array')
flats = initial_array.flatten()
j = -np.sort(-flats).reshape(4, 4)
print(j)
```

```python
print('a. numpyasg.txt text file')
filedata = np.genfromtxt('numpyasg.txt', delimiter=',')
filedata = filedata.astype('int32')
print(filedata)
print('b. numbers > 100')
nums = []
filedata = filedata.flatten()
for i in filedata:
    if(i > 100):
        nums.append(i)
print(nums)
print('c. numbers between 50 and 70')
nums_1 = []
for i in filedata:
    if(i > 50 and i < 70):
        nums_1.append(i)
print(nums_1)
```

<div class="alert alert-block alert-success"><b>Problem #8:</b>
    
5 x 5 System:<br>
<p>&nbsp;</p>
4a + b + 2c  3d + 5e = -16<br>
-3a + 3b  c + 4d  2e = 20<br>
-a + 2b + 5c + d + 3e = -4<br>
5a + 4b + 3c  d + 2e = -10<br>
a  2b + 3c  4d + 5e = 3<br>
<p>&nbsp;</p>
a. Initialize a 5 x 5 array named A that contains the coefficients of the system of equations found above.<br>
b. Initialize a 5 x 1 array named B that contains the constants of the system of equations found above.<br>
c. Find the determinant of array A.<br>
d. Find the inverse of array A.<br>
e. Find the product of array A and the inverse of array A.<br>
f. Find the product of the inverse of array A and array A.<br>
g. Solve the system above using matrix multiplication.<br>
h. Solve the system above using "numpy.linalg.solve()" ... see the link below.

[numpy.linalg.solve](https://numpy.org/doc/stable/reference/generated/numpy.linalg.solve.html)
<p>&nbsp;</p>
<center><u>Desired Output</u></center><br>



</div>

```python
print('a. Array A')
coefficients = np.array([[4, 1, 2, -3, 5], [-3, 3, -1, 4, -2],[-1, 2, 5, 1, 3], [5, 4, 3, -1, 2], [1, -2, 3, -4, 5]])
print(coefficients)
print('b. Array B')
constants = np.array([[-16], [20], [-4], [-10], [3]])
print(constants)
print('c. det_A')
print(np.linalg.det(coefficients))
print('d. inv_a')
inv_a = np.linalg.inv(coefficients)
print(inv_a)
print('e. A times inv_a')
prod = np.matmul(coefficients, inv_a)
print(prod)
print('f. inv_A times A')
prod_2 = np.matmul(inv_a, coefficients)
print(prod_2)
print('g. x = inv_a times b')
prod_3 = np.matmul(inv_a, constants)
print(prod_3)
print('h. Use np.linalg.solve() to solve the system')
print(np.linalg.solve(prod, prod_3))
```
