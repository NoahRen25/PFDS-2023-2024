**Use your function to find an answer to the following question:**<br>
    
There is a young man named Prathik who plans on attending a prestigious university. He has calculated that he will have to save $750,000 in the next two years to not only pay for 4 years of tuition at such a university, but to also fund the lavish lifestyle that he plans on living once he gets there. Fortunately for Prathik he currently has two sources of income -- first, he is fashion model for upscale men's magazines and second, he trains Navy Seals in close quarters combat.  From these two revenue sources, Prathik has calculated that for the next two years, each month he will be able to deposit \\$30,000 into an annuity due where he plans on earning 5.85% annual interest. (Note: He plans on saving the rest of his earnings to fund a cryptocurrency trading exchange that he plans on starting once he graduates from college and to buy a new Tesla upon his graduation.)<br>
    
Will Prathik have enough money in the annuity due at the end of two years to be able to completely fund his college plans or will he have to take on additional work in his university's Psychology labs cleaning out dog, rabbit and rodent kennels as part of a work-study program? Provide aesthetically formatted, detailed output to support you answer.

```python
def anuity_due(deposit, interest, periods):
    interest /= 1200
    eq_math = (((1 + interest)**(periods+1)) - 1) / interest
    s = deposit*eq_math - deposit
    return s
anuity = anuity_due(30000, 5.85, 24)
print(anuity)
test = ''
test_2 = ''
if(anuity >750000):
    test = 'more'
else: 
    test = 'less'
    test_2 = 'not '
print(f'Prathik will end up with ${round(anuity, 2)}. That is {test} than $750,000, so Prathik will {test_2}have enough money '
     f'to fund his lavish lifestyle and college tuition. at the end of his two-year anuity.')
```
