<b><font size="5">1. What Are Strings?</font></b>
</div>



<b>Unicode</b>, formally the <b>Unicode Standard</b>, is an [information technology standard](https://en.wikipedia.org/wiki/Information_technology) for the consistent [encoding](https://en.wikipedia.org/wiki/Character_encoding),representation, and handling of [text](https://en.wikipedia.org/wiki/Character_(computing)) expressed in most of the world's [writing systems](https://en.wikipedia.org/wiki/Writing_system). The standard, which is maintained by the [Unicode Consortium](https://en.wikipedia.org/wiki/Unicode_Consortium), defines 144,762 characters covering 159 modern and historic [scripts](https://en.wikipedia.org/wiki/Script_(Unicode)), as well as symbols, [emoji](https://en.wikipedia.org/wiki/Emoji), and non-visual control and formatting codes.<br>

[**Source:**](https://en.wikipedia.org/wiki/Unicode)

## Single quotes work

```python
single_quote = 'This is a valid string.'
print(single_quote)
```

## Double quotes work

```python
double_quote = "This is also a valid string."
print(double_quote)
```

```python
double_inside_single = ('The quote "when life gives you lemons, make lemonade" is credited' 
                         ' to Elbert Hubbard.')
print(double_inside_single)
```

```python
single_inside_double = "It's pure joy coding in Python."
print(single_inside_double)
```

<a class="anchor" id="3b"></a>
<div class="alert alert-block alert-danger">
<b><font size="4">b. String Concatenation</font></b>
</div>

```python
# First string concatenation example

string_1 = "Some LAUNCH classes "
string_2 = "are held at Concurrency."
first_new_string = string_1 + string_2
print(first_new_string)
```

```python
# Second string concatenation example

string_1 = "Some LAUNCH classes"
string_2 = " are held at Concurrency."
second_new_string = string_1 + string_2
print(second_new_string)
```

```python
# Third string concatenation example ... an example of what NOT to do

string_1 = "Some LAUNCH classes"
string_2 = "are held at Concurrency."
third_new_string = string_1 + string_2
print(third_new_string)
```
