<a class="anchor" id="1"></a>
<div class="alert alert-block alert-info">
<b><font size="4">1. append()</font></b></div>

- Adds an element at the end of the list

```python
sports = ['football', 'basketball', 'baseball']
print(sports)

sports.append('tennis')
print(sports)
```

<a class="anchor" id="2"></a>
<div class="alert alert-block alert-info">
<b><font size="4">2. clear()</font></b></div>

- Removes all the elements from the list

```python
cities = ['Chicago', 'New York City', 'Boston', 'Los Angeles']
print(cities)

cities.clear()
print(cities)
```

<a class="anchor" id="3"></a>
<div class="alert alert-block alert-info">
<b><font size="4">3. copy()</font></b></div>

- Returns a copy of the list

```python
fruits = ['apple', 'banana', 'cherry', 'orange']
print(fruits)

x = fruits.copy()
print(x)
```

<a class="anchor" id="4"></a>
<div class="alert alert-block alert-info">
<b><font size="4">4. count()</font></b></div>

- Returns the number of elements with the specified value

```python
states = ['Wisconsin', 'Minnesota', 'Illinois', 'Nebraska', 'Indiana', 'Iowa', 'Illinois']
print(states)

x = states.count("Illinois")
print(x)
```

<a class="anchor" id="5"></a>
<div class="alert alert-block alert-info">
<b><font size="4">5. extend()</font></b></div>

- Add the elements of a list (or any iterable), to the end of the current list

```python
fruits = ['apple', 'banana', 'cherry', 'peach', 'orange']
print(fruits)

cars = ['Ford', 'BMW', 'Volvo', 'Honda']
print(cars)

print()

fruits.extend(cars)
print(fruits)

cars.extend(fruits)
print(cars)
```

<a class="anchor" id="6"></a>
<div class="alert alert-block alert-info">
<b><font size="4">6. index()</font></b></div>

- Returns the index of the first element with the specified value

```python
subjects = ['math', 'reading', 'science', 'social studies', 'math', 'PE', 'english', 'science', 'computer science', 
            'business']
x = subjects.index('science')
print(x)
```

<a class="anchor" id="7"></a>
<div class="alert alert-block alert-info">
<b><font size="4">7. insert()</font></b></div>

- Adds an element at the specified position

```python
planets = ['Earth', 'Mars', 'Saturn']
print(planets)

planets.insert(2, 'Jupiter')
print(planets)
```

<a class="anchor" id="8"></a>
<div class="alert alert-block alert-info">
<b><font size="4">8. pop()</font></b></div>

- 	Removes the element at the specified position

```python
numbers = ['five', 'nine', 'two', 'seven', 'six']
print(numbers)

numbers.pop(3)
print(numbers)
```

```python
numbers = ['five', 'nine', 'two', 'seven', 'six']
print(numbers)

numbers.pop()
print(numbers)
```

<a class="anchor" id="9"></a>
<div class="alert alert-block alert-info">
<b><font size="4">9. remove()</font></b></div>

- Removes the item with the specified value

```python
veggies = ['tomato', 'broccoli', 'zucchini', 'banana', 'asparagus']
print(veggies)

veggies.remove("banana")
print(veggies)
```

<a class="anchor" id="10"></a>
<div class="alert alert-block alert-info">
<b><font size="4">10. reverse()</font></b></div>

- Reverses the order of the list

```python
instruments =['guitar', 'piano', 'violin', 'harp']
print(instruments)

instruments.reverse()
print(instruments)
```

<a class="anchor" id="11"></a>
<div class="alert alert-block alert-info">
<b><font size="4">11. sort()</font></b></div>

- Sorts the list

```python
directions = ['north', 'south', 'east', 'west']
print(directions)

directions.sort()
print(directions)
```

```python
numbers = [10, 16.3, 2000, -20.7]
print(numbers)

numbers.sort()
print(numbers)
```
