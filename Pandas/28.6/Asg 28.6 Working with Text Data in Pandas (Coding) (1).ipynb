<div class="alert alert-block alert-success"><b>Question 10:</b><br>
    
Convert the <b>Price</b> Series of the <b>prices</b> DataFrame from a string to a float.   

```python
prices['Price'].str[3:].astype(float)
```

<div class="alert alert-block alert-success"><b>Question 11:</b><br>
Create a new column named <b>'Revenue'</b>. The <b>'Revenue'</b> column can be derived by using the following formula:<br>

Revenue = Price * Quantity Sold

```python
prices['Revenue'] = prices['Price'].str[3:].astype(float) * prices['Quantity Sold']
prices['Revenue']
```

<div class="alert alert-block alert-success"><b>Question 12</b><br>
