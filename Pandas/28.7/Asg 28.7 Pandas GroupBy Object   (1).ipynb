<div class="alert alert-block alert-success"><b>Question 2:</b>
  
Calculate the mean <b>Income</b> across the entire dataset.

```python
customers['Income'].mean()
```

<div class="alert alert-block alert-success"><b>Question 16:</b>
  
Calculate the mean <b>Income</b> just for Single people.

```python
customers[(customers['Marital_Status'] == 'Single')]['Income'].mean()
```

<div class="alert alert-block alert-success"><b>Question 3:</b>
  
Calculate the mean <b>Income</b> for each <b>Marital_Status</b>.
    
<p>&nbsp;</p>
    
Documentation for [**'groupby'**](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.groupby.html)

```python
groups = customers.groupby('Marital_Status')
groups.get_group('Single')['Income'].mean()
#b
groups.get_group('Married')['Income'].mean()
#c
groups.get_group('Divorced')['Income'].mean()
#d
groups.get_group('Together')['Income'].mean()
groups.get_group('YOLO')['Income'].mean()
groups.get_group('Absurd')['Income'].mean()
groups.get_group('Alone')['Income'].mean()
groups.get_group('Widow')['Income'].mean()
```

<div class="alert alert-block alert-success"><b>Question 4:</b>
  
a. Calculate the maximum <b>Income</b> for each <b>Marital_Status</b>.<br>

b. Calculate the minimum <b>Income</b> for each <b>Marital_Status.</b><br>

```python
groups.get_group('Single')['Income'].max()
groups.get_group('Married')['Income'].max()
groups.get_group('Divorced')['Income'].max()
groups.get_group('Together')['Income'].max()
groups.get_group('YOLO')['Income'].max()
groups.get_group('Alone')['Income'].max()
groups.get_group('Absurd')['Income'].max()
groups.get_group('Widow')['Income'].max()
#b
print('\n')
groups.get_group('Single')['Income'].min()
groups.get_group('Married')['Income'].min()
groups.get_group('Divorced')['Income'].min()
groups.get_group('Together')['Income'].min()
groups.get_group('YOLO')['Income'].min()
groups.get_group('Alone')['Income'].min()
groups.get_group('Absurd')['Income'].min()
groups.get_group('Widow')['Income'].min()
```

<div class="alert alert-block alert-success"><b>Question 5:</b>

Calculate the count, mean, median, maximum, and minimum for the <b>Income</b> Series for each <b>Marital Status</b> ... i.e., apply multiple aggregation functions simultaneously.

```python
customers.groupby('Marital_Status').agg({'Income':['min', 'mean', 'median', 'max']})
```

<div class="alert alert-block alert-success"><b>Question 8:</b>

Calculate the mean for all numeric columns for each level of <b>Education</b>.

```python
groups = customers.groupby('Education')
groups.get_group('Graduation').select_dtypes(include='number').mean()
groups.get_group('PhD').select_dtypes(include='number').mean()
groups.get_group('Master').select_dtypes(include='number').mean()
groups.get_group('Basic').select_dtypes(include='number').mean()
groups.get_group('2n Cycle').select_dtypes(include='number').mean()
```
