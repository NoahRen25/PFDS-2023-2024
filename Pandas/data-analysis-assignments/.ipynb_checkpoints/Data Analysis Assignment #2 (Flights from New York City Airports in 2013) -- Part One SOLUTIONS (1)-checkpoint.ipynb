<div class="alert alert-block alert-danger">
<b>Question 10:</b>

Which carrier had the most flights from a New York City airport to Milwaukee's General Mitchell International Airport (MKE) in 2013?
    
</div>

```python
flights_to_MKE = flights.loc[(flights['dest'] == "MKE")]

flights_to_MKE.head()
```

```python
flights_to_MKE.carrier.max()

flights_to_MKE.carrier.value_counts()
```

The carrier that had the most flights from a New York City airport to Milwaukee's General Mitchell International Airport (MKE) in 2013 was WN (i.e., Southwest Airlines). WN (Southwest) had 128 flights to MKE.
