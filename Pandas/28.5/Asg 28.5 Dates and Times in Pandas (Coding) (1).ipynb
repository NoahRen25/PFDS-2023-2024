Create individual timestamps by parsing the following formatted string dates:

```python
from dateutil import parser
date_1 = parser.parse('1st of January, 2022')
date_2 = parser.parse('February 14th, 2022')
date_3 = parser.parse('4th of July, 2022')
date_1
date_2
date_3
```
