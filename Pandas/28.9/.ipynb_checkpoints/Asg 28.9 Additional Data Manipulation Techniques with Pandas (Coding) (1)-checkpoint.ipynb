<p>&nbsp;<p> 
    a. Read in the dataset <b>cereal.csv</b> and store the results in a DataFrame named <b>cereal</b>.<br>
<p>&nbsp;<p>  
b. Use the Pandas agg() method to find the median for every column that has a numeric variable in the cereal DataFrame.<br>
<p>&nbsp;<p>  
c. Use the Pandas agg() method to find the highest rating in the cereal DataFrame.<br>
<p>&nbsp;<p> 
d. Use the Pandas agg() method to find the standard deviation of the calories, sodium, and sugars columns of the cereal DataFrame.<br>
<p>&nbsp;<p> 
e. Use the Pandas agg() method to find the mean, median, standard deviation, maximum value, and minimum value of the sodium column of the cereal DataFrame.<br>
<p>&nbsp;<p>  
f. Use the Pandas agg() method to find the mean, median, standard deviation, maximum value, and minimum value of the calories, fiber, and potass columns of the cereal DataFrame.<br>
</div>

```python
cereal = pd.read_csv('cereal.csv')
cereal.agg('median', numeric_only = True)
cereal['rating'].agg('max', numeric_only = True)
cereal[['calories', 'sodium', 'sugars']].agg('std', numeric_only=True)
cereal['sodium'].agg(['mean', 'median', 'std', 'max', 'min'])
cereal[['calories', 'fiber', 'potass']].agg(['mean', 'median', 'std', 'max', 'min'])
```

<div class="alert alert-block alert-success"><b>Task 3:</b>
 
a. Read in the dataset <b>Big_Ten_East.xlsx</b> and store the results in a DataFrame named <b>east</b>. Use the School column as the index.<br>
<p>&nbsp;</p> 
b. Read in the dataset <b>Big_Ten_West.xlsx</b> and store the results in a DataFrame named <b>west</b>. Use the School column as the index.<br>
<p>&nbsp;</p> 
c. Use the Pandas concat() method to combine the two DataFrames into a single DataFrame named <b>big_ten</b>.<br>
<p>&nbsp;</p>
d. Use the Pandas assign() method to create a new DataFrame named <b>big_ten_pcts</b> where a column for winning percentage is added to the <b>big_ten</b> DataFrame. Name this column Pct. <br>
<p>&nbsp;</p>
e. Sort the <b>big_ten_pcts</b> DataFrame in descending order by Pct.<br>
<p>&nbsp;</p>
f. Use the Pandas filter() method to retrieve the Wins, Losses, and Pct columns from the <b>big_ten_pcts</b> DataFrame.<br>
</div>

```python
east = pd.read_csv('Big_Ten_East.csv').set_index('School')
west = pd.read_csv('Big_Ten_West.csv').set_index('School')
east
west
big_ten = pd.concat([east, west])
big_ten = big_ten.assign(big_ten_pcts = big_ten['Wins'] / (big_ten['Wins'] + big_ten['Losses']))
big_ten['big_ten_pcts'].sort_values(ascending=True)
big_ten.filter(['Wins', 'Losses', 'big_ten_pcts'])
```

<div class="alert alert-block alert-success"><b>Task 4:</b>
    
Use the Pandas query() method to find the following subsets of the final (i.e., sorted) <b>big_ten_pcts</b> DataFrame that you created in Task #3.<br>

c. All Big 10 teams from Indiana or teams that have the nickname Badgers.

```python
big_ten.query('Wins >= 6')
big_ten.query("State == 'Michigan' or Losses > 5")
big_ten.query("State == 'Indiana' or Nickname == 'Badgers'")
big_ten.query("(State == 'Illnois' and Losses > 5) or State == 'Michigan'")
```

```python
students = pd.DataFrame({'Student Names' : ['Jenny', 'Singh', 'Charles', 'Richard', 'Veena'],
                       'Category' : ['Online', 'Offline', 'Offline', 'Offline', 'Online'],
                       'Gender' : ['Female', 'Male', 'Male', 'Male', 'Female'],
                      'Courses': ['Java', 'Spark', 'PySpark','Hadoop','C'],
                       'Fee': [15000, 17000, 27000, 29000, 12000],
                       'Discount': [1100, 800, 1000, 1600, 600]})

students

students.pivot_table(index=['Gender'])
```
