## <span style="color:blue"><b>1. THE ASSIGN METHOD ADDS NEW VARIABLES TO A DATAFRAME</b></span>

```python
# Initialize list of lists

students = [['John', 43, 48, 47, 50, 92, 96],
            ['Mary', 48, 46, 50, 44, 93, 95],
            ['Ed', 36, 41, 38, 45, 86, 82],
            ['Tom', 40, 50, 40, 50, 80, 100],
            ['Sally', 50, 46, 48, 50, 99, 97]]

# Create the pandas DataFrame

grades = pd.DataFrame(students, columns = ['Name', 'Quiz_1', 'Quiz_2', 'Quiz_3', 'Quiz_4', 'Test_1', 'Test_2'])

grades
```

```python
grades.assign(Quiz_Avg = (grades.Quiz_1 + grades.Quiz_2 + grades.Quiz_3 + grades.Quiz_4) / 4)
```

<a class="anchor" id="22"></a>
## <span style="color:blue"><b>2. SYNTAX OF PANDAS ASSIGN</b></span>

```python
grades.assign(Quiz_Avg = (grades.Quiz_1 + grades.Quiz_2 + grades.Quiz_3 + grades.Quiz_4) / 4, 
              Test_Avg = (grades.Test_1 + grades.Test_2) / 2)
```

<a class="anchor" id="24"></a>
## <span style="color:blue"><b>4. THE OUTPUT OF PANDAS ASSIGN</b></span>

#### Pandas assign outputs a new DataFrame
- Contains all of the old variables
- Also contains the new variables

## <span style="color:blue"><b>1. THE FILTER METHOD RETRIEVES COLUMNS FROM A DATAFRAME</b></span>

```python
student_info = pd.read_csv("student_info.csv")

student_info
```

```python
student_info.filter(['Superpower'])
```

#### NOTE: PANDAS FILTER TECHNICALLY CAN RETRIEVE ROWS TOO

- Technically, the filter method can retrieve rows too
- Using filter to retrieve rows is discouraged because Pandas has better tools for retrieving rows like … query, loc, iloc
- Therefore it is recommended that filter be used for columns, just for the sake of simplicity. 

<a class="anchor" id="42"></a>
## <span style="color:blue"><b>2. SYNTAX OF PANDAS FILTER</b></span>

```python
student_info.filter(['Handed', 'Favorite_Music', 'Grade'])
```

```python
student_info.filter(['Gender', 'Age', 'Superpower'])
```

```python
# Initialize list of lists

students = [['John', 43, 48, 47, 50, 92, 96],
            ['Mary', 48, 46, 50, 44, 93, 95],
            ['Ed', 36, 41, 38, 45, 86, 82],
            ['Tom', 40, 50, 40, 50, 80, 100],
            ['Sally', 50, 46, 48, 50, 99, 97]]

# Create the pandas DataFrame

grades = pd.DataFrame(students, columns = ['Name', 'Quiz_1', 'Quiz_2', 'Quiz_3', 'Quiz_4', 'Test_1', 'Test_2'])

grades
```

```python
grades.melt(id_vars = 'Name', 
            value_vars = ['Quiz_1', 'Quiz_2', 'Quiz_3', 'Quiz_4', 'Test_1', 'Test_2'])
```

<a class="anchor" id="61"></a>
## <span style="color:blue"><b>1. WHAT IS A PANDAS MERGE?</b></span>

- A merge combines records from 2 datasets

<a class="anchor" id="62"></a>
## <span style="color:blue"><b>2. SYNTAX OF PANDAS MERGE </b></span>

```python
data_1 = [[101, 'Green Bay', 'Packers'],
        [102, 'Chicago', 'Bears'],
        [103, 'Minnesota', 'Vikings'],
        [104, 'Detroit', 'Lions'],
        [105, 'Philadelphia', 'Eagles'],
        [106, 'Dallas', 'Cowboys']
       ]

teams = pd.DataFrame(data_1, columns = ['Team_ID', 'City', 'Nickname'])

teams

data_2 = [[101, 'Aaron', 'Rodgers'],
        [102, 'Justin', 'Fields'],
        [103, 'Kurt', 'Cousins'],
        [104, 'Jared', 'Goff'],
       ]

quarterbacks = pd.DataFrame(data_2, columns = ['Team_ID', 'First_Name', 'Last_Name'])

quarterbacks
```

```python
pd.merge(teams, quarterbacks, on = 'Team_ID', how = 'left')
```

```python
pd.merge(teams, quarterbacks, on = 'Team_ID', how = 'inner')
```

<a class="anchor" id="63"></a>
## <span style="color:blue"><b>3. THE PARAMETERS OF PANDAS MERGE</b></span>

<a class="anchor" id="64"></a>
## <span style="color:blue"><b>4. THE OUTPUT OF PANDAS MERGE</b></span>

- The output of pd.merge() is a DataFrame
  - Contains the merged data
  - Merged as specified in your syntax

<a class="anchor" id="72"></a>
## <span style="color:blue"><b>2. SYNTAX OF PANDAS PIVOT</b></span>

```python
stocks = pd.read_csv("stocks.txt")

stocks
```

```python
# Pivot was used to examine the closing trading price for each stock symbol over the trading window

stocks.pivot(index='symbol', columns='date', values='close')
```

**Note:**
- The index and column parameters are interchangeable. 
- If you want to reorganize so that the dates are used as the index and the stock symbols are my columns, you can just flip the parameters.

```python
stocks.pivot(index='date', columns='symbol', values='close')
```

## <span style="color:blue"><b>4. THE OUTPUT OF PANDAS QUERY</b></span>

#### THE OUTPUT OF PANDAS QUERY

- By default, query() will produce a new DataFrame
  - DataFrame that includes the rows that meet the specified condition
  - Remember: by default inplace = False
-If you set inplace = True, query() will directly modify the DataFrame you're operating on
