<div class="alert alert-block alert-success"><b>Question 5:</b><br>
    
Filter the <b>customers</b> DataFrame rows to show only customers with an <b>'Income'</b> of at least 80000 and a <b>'Marital_Status'</b> of Single.

```python
high_income = customers['Income'] > 80000
single = customers['Marital_Status'] == 'Single'
customers[high_income & single]
```

<div class="alert alert-block alert-success"><b>Question 6:</b><br>
    
Filter the <b>customers</b> DataFrame rows to show only customers with an <b>'Income'</b> greater than 80000, a <b>'Marital_Status'</b> of Divorced, and an '<b>'Education'</b> level of Graduation.

```python
high_income = customers['Income'] > 80000
divorced = customers['Marital_Status'] == 'Divorced'
grad = customers['Education'] == 'Graduation'

customers[high_income & divorced & grad]
```

<div class="alert alert-block alert-success"><b>Question 7:</b><br>
    
Filter the <b>customers</b> DataFrame rows to show only customers with a <b>'Marital_Status'</b> of Married or Together or an '<b>'Education'</b> level of PhD.

```python
PhD = customers['Education'] == 'PhD'
married = customers['Marital_Status'] == 'Married'
together = customers['Marital_Status'] == 'Together'
customers[PhD & (married | together)]
```

<div class="alert alert-block alert-success"><b>Question 8:</b><br>
    
Filter the <b>customers</b> DataFrame rows to show only customers with an <b>Income</b> between 40000 and 60000 or an '<b>'Education'</b> level of Master or PhD.

```python
lower_income = customers['Income'] > 40000
higher_income = customers['Income'] < 60000
masters = customers['Education'] == 'Master'
PhD = customers['Education'] == 'PhD'
customers[lower_income & higher_income & (masters | PhD)]
```

<div class="alert alert-block alert-success"><b>Question 9:</b><br>
    
Use the <b>"isin() method"</b> to filter the <b>customers</b> DataFrame to show only customers with a <b>Marital_Status</b> of Single, Married, or Together.<br>
<p>&nbsp;</p>
 
Documentation for [**'isin'**](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.isin.html)

```python
not_divorced = customers['Marital_Status'].isin(['Single', 'Married', 'Together'])
customers[not_divorced]
```
