<div class="alert alert-block alert-success"><b>Problem #1:</b>

Write a **well-documented Python program** to create a class representing a Circle. 
    
- Include methods to calculate the circle's area and circumference. 
- Prompt the user to input the radius of the circle and convert it to a floating-point number.

Test your program on a circle that has a radius of 10.
    <p>&nbsp;</p>
<u>Expected Output</u><br>
    

    
</div>

```python
import math


class Circle:
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return radius**2 * math.pi

    def circumference(self):
        return 2 * math.pi * radius


r = eval(input("Input the radius of the circle: "))
circle = Circle(r)
print(f'Area of the circle: {circle.area()}')
print(f'Perimter of the circle: {circle.circumference()}')
```

<div class="alert alert-block alert-success"><b>Problem #2:</b> 

Write a **well-documented Python program** to create a calculator class. Include methods for the basic arithmetic operations ... addition. subtraction, multiplication, and division.  Also include methods for finding powers, square roots, and logarithms.
    
Test your program on the following calculations:
- 10 + 15
- 50 - 30
- 35 * 4
- 220 / 11
- 500 / 0
- 3 ^ 4
- $\sqrt{64}$
- $\sqrt{-100}$
- ln 15
- ln 0
- log 100
- log (-50)
    
<u>Expected Output</u>
    

    
</div>

```python
import math


class Calculator:
    def __init__(self):
        pass

    def addition(self, number, i):
        return number + i

    def subtraction(self, number, i):
        return number - i

    def multiplication(self, number, i):
        return number

    def division(self, number, i):
        if (i != 0):
            return number / i
        else:
            return 'Cannot divide by zero'

    def exponent(self, number, i):
        return number**i

    def square_root(self, number):
        if number > 0:
            return math.sqrt(number)
        else:
            return 'is not defined'

    def natural_log(self, number):
        if (number > 0):
            return math.log(number)
        else:
            return 'Is not defined'

    def common_log(self, number):
        if (number > 0):
            return math.log10(number)
        else:
            return 'Is not defined'


calculate = Calculator()
print(calculate.addition(10, 15))
print(calculate.subtraction(50, 30))
print(calculate.multiplication(35, 4))
print(calculate.division(220, 11))
print(calculate.division(500, 0))
print(calculate.exponent(3, 4))
print(calculate.square_root(64))
print(calculate.square_root(-100))
print(calculate.natural_log(15))
print(calculate.natural_log(0))
print(calculate.common_log(100))
print(calculate.common_log(-15))
```
