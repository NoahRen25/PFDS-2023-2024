<div class="alert alert-block alert-success"><b>Question #2:</b> 
    
Tom had to take 4 tests during his first-term computer science class.  His scores on those tests were 87 out of 100, 72 out of 75, 50 out of 50, and 71 out of 80.
<p>&nbsp;</p>    
Write the proper code to do the following:
<p>&nbsp;</p>
a. Create two variables -- one called <b>points_earned</b> and the other called <b>points_possible</b> and assign the points that Tom earned on the first test to the variable <b>points_earned</b> and assign the points possible on the first test to the variable <b>points_possible</b>. 
<p>&nbsp;</p> 
b. Update the two variables after each test.
<p>&nbsp;</p> 
c. After the final test, display Tom's overall percent on the tests ...<br>
<p>&nbsp;</p>    
i. to the nearest integer (be careful here!) in an output statement that says ... "Tom's overall average on the 4 computer science tests was a ___ %."  In the blank include Tom's overall percent that you calculated.<br>
    
ii. rounded to 3 places in an output statement that says ... "Tom's overall average on the 4 computer science tests was a ___ %."  In the blank include Tom's overall percent that you calculated.
<p>&nbsp;</p> 
Note: Include appropriate comments in your could.
</div>

```python
points_earned = 87
points_possible = 100
```

```python
points_earned2 = 72
points_possible2 = 75
points_earned3 = 50
points_possible3 = 50
points_earned4 = 71
points_possible4 = 80
```

```python
percentage = (points_earned + points_earned2 + points_earned3 + points_earned4) / (points_possible + points_possible2 + points_possible3 + points_possible4) * 100
print(percentage)
```

```python
casted_percent = int(percentage+0.5)
print(f'Toms overall average on the 4 computer science test was a {casted_percent} %')
```

```python
str = str(round(percentage, 3))
print('Toms overall average on the 4 computer science test was a ' + str + '%')
```

<div class="alert alert-block alert-success"><b>Question #4:</b> 
The formula for computing the final amount if one is earning compound interest is given on Wikipedia as


<p>&nbsp;</p>   
a. Write a Python program that assigns the principal amount of 25000 to variable P, assigns to n the value 12, and assigns to r the interest rate of 6% (0.06). Then have the program prompt the user for the number of years, t, that the money will be compounded for. 

b. Calculate and print the final amount if the user answers that the number of years is t = 15.
  
</div>

```python
P = 25000
n = 12
r = 0.06
t = int(input("Enter the number of years your money will compound"))
final_amt = P*((1 + (r/n))**(n*t))
print(round(final_amt,2))
```

<p>&nbsp;</p>     
b. Test your code for a person named Mary who has lived for 536210363 seconds

```python
seconds_alive = int(input('Please enter the seconds you have been alive:'))
years_alive = int(seconds_alive / 31536000)
seconds_left = seconds_alive - 31536000*years_alive
days_alive = int(seconds_left / 86400)
seconds_left = seconds_left - 86400 * days_alive
hours_alive = int(seconds_left / 3600)
seconds_left = seconds_left - 3600 * hours_alive
minutes_alive = int(seconds_left / 60)
seconds_left = seconds_left - minutes_alive * 60
print(f'You have been alive for {years_alive} years, {days_alive} days, {hours_alive} hours,  {minutes_alive} minutes, {seconds_left} seconds') 
```
